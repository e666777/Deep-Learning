# 一元线性回归

```python
import keras
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## 生成数据

```python
x = np.linspace(0, 10, 300)
y = 2*x + 5 + np.random.randn(300)
print(x)
print(y)
```

```python
#可视化生成的数据
plt.scatter(x,y)
```

## 使用梯度下降法计算参数a、b

## 构建模型

```python
from keras import layers

model = keras.Sequential()#初始模型：顺序模型
model.add(layers.Dense(1,input_dim=1))#输入和输出模型都是一维模型
model.summary()
```

## 编译模型

```python
model.compile(optimizer="adam",loss = "mse")#adam作为优化算法，优化目标loss为均方差
```

## 训练模型

```python
model.fit(x,y,epochs=5000)#模型训练五千次
```

## 输出结果

```python
plt.scatter(x,y,c="r")
plt.plot(x,model.predict(x))
model.predict([150])
```
